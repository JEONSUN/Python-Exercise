```python
# 라이브러리 호출
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
plt.style.use('classic')
```

```python
# plot 실행 바로 확인
%matplotlib inline
```

```python
# linspace : 구간 0부터 10까지 동일한 간격의 점 100개로 구성된 벡터 생성
x = np.linspace(0,10,100)
x
```

```python
y = x**2
y
```

```python
# plot 그리기
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.show()
```

```python
# plot 그리기
plt.plot(x, np.sin(x))
plt.plot(x,np.cos(x))
plt.show()
```

```python
# 선 변경
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x),'--')

plt.show()
```

```python
# plot 추가
plt.subplot(1,2,1)
plt.plot(x,y)

plt.subplot(1,2,2)
plt.plot(y,x)

plt.show()
```

### subplot 사용

```python
# 빈 캠버스 추가
plt.figure()


# 두 개의 plot중 첫 번째 pot을 생성하고 현재 축을 설정
plt.subplot(2,1,1)
plt.plot(x, np.sin(x))

# 두 번째 패널을 생성하고 축 설정
plt.subplot(2,1,2)
plt.plot(x, np.cos(x))

plt.show()
```

```python
# 간단히 축약
# ax는 두개의
fig, ax = plt.subplots(2)

# 적절한 객체에서 plot() 메소드를 호출
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))

plt.show()
```

### 색 설정

```python
# 선의 색을 파란색으로 직접 설정
plt.plot(x, np.sin(x - 0), color = 'blue')

plt.show()
```

```python
# 색 코드로 선의 색 설정(rgbcmyk)
plt.plot(x, np.sin(x - 1), color = 'g')

plt.show()
```

```python
# 0과 1 사이로 회색조 지정
plt.plot(x, np.sin(x - 2), color = '0.3')
plt.show()
```

### 선 설정

```python
plt.plot(x, x+0, linestyle = 'solid')
plt.plot(x, x+1, linestyle = 'dashed')
plt.plot(x, x+2, linestyle = 'dashdot')
plt.plot(x, x+3, linestyle = 'dotted')

plt.show()
```

```python
### 축약형

plt.plot(x, x+4 , linestyle = '-')
plt.plot(x, x+5 , linestyle = '--')
plt.plot(x, x+6 , linestyle = '-.')
plt.plot(x, x+7 , linestyle = ':')

plt.show()
```

```python
# 색과 선을 합쳐서 간결하게 표현
plt.plot(x,x+0,'-g')
plt.plot(x,x+1,'--c')
plt.plot(x,x+2,'-.k')
plt.plot(x,x+3,':r')

plt.show()
```

### 그래프 조정하기 : 축

```python
# 축 경계 선정
plt.plot(x, np.sin(x))

plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
plt.show()
```

```python
# x축과 y축을 역으로 표시
plt.plot(x, np.sin(x))

plt.xlim(10,0)
plt.ylim(1.5,-1.5)

plt.show()
```

```python
# 축경계를 한번에 호출
plt.plot(x, np.sin(x))
plt.axis([-1,11,-1.5,1.5])
plt.show()
```

```python
# 그래프 주변 경계를 자동으로 더 밀착시킬 수 있음.
plt.plot(x, np.sin(x))
plt.axis('tight')
plt.show()
```

```python
# x축과 y축의 단위가 똑같게 설정
plt.plot(x, np.sin(x))
plt.axis('equal')
plt.show()
```

```python
# plot에 레이블 붙이기

plt.plot(x, np.sin(x))
# 제목
plt.title('Sin Curve')
# x축
plt.xlabel('x')
# y축
plt.ylabel('sin(x)')

plt.show()
```

```python
# 범례추가 : label 키워드를 사용하여 각 라인의 레이블을 지정
plt.plot(x , np.sin(x), '-g',label = 'sin(x)')
plt.plot(x, np.cos(x), ':b', label = 'cos(x)')
plt.axis('equal')

# legend를 사용하여 범례를 지정하고 매칭
plt.legend()
plt.show()
```

```python
# 한번에 표현이 가능함 
ax = plt.axes()

ax.plot(x, np.sin(x))

# set을 사용해 속성 설정
ax.set(xlim = (0,10), ylim = (-2,2),
       xlabel = 'x', ylabel = 'sin(x)',
       title = 'Simple Plot')
plt.show()
```

### 산점도

```python
x = np.linspace(0,10,30)
y = np.sin(x)

plt.plot(x,y,'o',color = 'black')
plt.axis('equal')
plt.show()
```

```python
# - : 선, o : 기호, k : 검정색
plt.plot(x,y,'-ok')
plt.show()
```

```python
# 마커 색, 크기 조절
plt.plot(x,y, '-p', color = 'blue', 
        markersize = 10, linewidth = 4,
        markerfacecolor ='white',
        markeredgecolor=  'gray',
        markeredgewidth = 2)

plt.ylim(-1.2,1.2)
plt.show()
```

```python
# plt.scatter
plt.scatter(x,y, marker = 'o')
plt.show()
```

plt.scatter는 plt.plot과 달리 각 점의 속성을 개별적으로 제어

대신 plt.scatter는 각 점을 개별적으로 구성하는 추가 작업을 해야함.

대용량 데이터는 plt.plot을 사용하는 것이 낫다.

```python
# 난수 생성
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)

# 색상과 크기 조절
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

# scatter 플롯
plt.scatter(x, y, c = colors, s = sizes, alpha = 0.3,
           cmap = 'viridis')

# 색상 척도 표시
plt.colorbar()
plt.show()
```
