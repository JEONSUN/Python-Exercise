```python
res = requests.get(url)
res.raise_for_status() # 문제시 프로그램 종료
soup = BeautifulSoup(res.text, "lxml") 

posts = soup.find_all("li", attrs={"class":"bx _svp_item"})
```

```python
for post in posts:
    post_title = post.find("a", attrs = {'class' : 'api_txt_lines total_tit'}).get_text()
    print("제목 :", post_title)
    post_link = post.find("a", attrs = {'class' : 'api_txt_lines total_tit'})['href']
    
    print('link :',post_link)
    print("-"*50)
```

```python
url_list = []
title_list =[]

query = input("검색어를 입력합시다: " )
# quote 함수를 사용하여 url 인코딩
url = "https://search.naver.com/search.naver?where=view&sm=tab_viw.blog&query=" + quote(query)


browser.get(url)
browser.implicitly_wait(5)


browser.find_element_by_xpath('//*[@id="snb"]/div[1]/div/div[1]/a[2]').click()
browser.implicitly_wait(5)
xxpath = '//*[@id="sp_blog_1"]/div/div/a'
for i in range(1,15) :
    # 블로그 들어가기
    print('-'*30)
    print("{}번 째 블로그 진행중".format(i))
    browser.find_element_by_xpath(xxpath).click()
    browser.implicitly_wait(5)
    browser.switch_to.window(browser.window_handles[-1])
    browser.implicitly_wait(5)
    

    # 현재 탭 닫기
    browser.close()
    browser.implicitly_wait(5)

    # 맨 처음 탭으로 변경(0번 탭)
    browser.switch_to.window(browser.window_handles[0])
    browser.implicitly_wait(5)
    z = str(i)
    if i > 10 :
        xxpath = xxpath.replace(xxpath[17]+xxpath[18],z)
    else :
        xxpath = xxpath.replace(xxpath[17],z)
```

```python
# 빈창 만들기
options = webdriver.ChromeOptions()
options.add_argument('--start-maximized')

driver = webdriver.Chrome('chromedriver', options = options)
article_list = []
url_list = []
content_list = ""
text = "아이오닉5 시승기"
 
for i in range(1, 2):  # 1~2페이지까지의 블로그 내용을 읽어옴
    url = 'https://section.blog.naver.com/Search/Post.nhn?pageNo='+ str(i) + '&rangeType=ALL&orderBy=sim&keyword=' + text
    driver.get(url)
    time.sleep(0.5)
 
    for j in range(1, 4): # 각 블로그 주소 저장
        titles = driver.find_element_by_xpath('/html/body/ui-view/div/main/div/div/section/div[2]/div['+str(j)+']/div/div[1]/div[1]/a[1]')
        title = titles.get_attribute('href')
        url_list.append(title)
        article = titles.get_attribute('title')
        article_list.append(article)
print("url 수집 끝, 해당 url 데이터 크롤링")
 
for url in url_list: # 수집한 url 만큼 반복
    driver.get(url) # 해당 url로 이동
 
    driver.switch_to.frame('mainFrame')
    overlays = ".se-component.se-text.se-l-default" # 내용 크롤링
    contents = driver.find_elements_by_css_selector(overlays)
 
    for content in contents:
        content_list = content_list + content.text # content_list 라는 값에 + 하면서 점점 누적
        

print("크롤링도 끝")
driver.quit()
```
