```python
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
```

```python
# 빈창 만들기
options = webdriver.ChromeOptions()
options.add_argument('--start-maximized')

driver = webdriver.Chrome('chromedriver', options = options)
```

### 수집할 사이트 정보 추출

```python
# 사이트 링크 입력
url = "https://www.teamblind.com/kr/search/%EC%A0%84%EA%B8%B0%EC%9E%90%EB%8F%99%EC%B0%A8"
driver.get(url)

# 카테고리를 입력해야함
driver.find_element_by_xpath('//*[@id="m_channel"]/option[11]').click()

body = driver.find_element_by_tag_name('body') # body를 스크롤하기 위해 tagname이 body로 되어있는것을 추출합니다.
num_of_pagedowns = 300 

SCROLL_PAUSE_SEC = 1

# 스크롤 높이 가져옴
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # 끝까지 스크롤 다운
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # 1초 대기
    time.sleep(SCROLL_PAUSE_SEC)

    # 스크롤 다운 후 스크롤 높이 다시 가져옴
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

print("페이지 읽기 종료")
soup = BeautifulSoup(driver.page_source, 'html.parser')
driver.close()
```

### 원하는 정보 추출

- 링크와 포스트 명

```python
anchors = soup.find_all('a', {'class': '', 'href': True})
result = []
text_list = []
for i in anchors :
    result.append(i['href'])
    text_list.append(i.text)
```

```python
link_text = pd.DataFrame({"link" : result, "text" : text_list})
link_text
```

```python
url = 'https://www.teamblind.com'
url_list = []
text_list = []
for i in range(1,len(link_text)) :
    if link_text['text'][i] == "자동차" :
        url_list.append(url + str(link_text['link'][i+1]))
        text_list.append(str(link_text['text'][i+1]))
        if link_text['link'][i].find('post') == 0 :
            pass
```

```python
url_df = pd.DataFrame({"url": url_list,"text" : text_list})
#url_df.to_excel("블라인드(전기자동차 - 자동차)링크.xlsx",index = None)
```

### 해당 링크를 통해 포스트명, 본문, 댓글 추출

```python
# 제목, 내용, 댓글 추출
title_list = []
main_list = []
comment_list = []
options = webdriver.ChromeOptions()
options.add_argument('--start-maximized')
driver = webdriver.Chrome('chromedriver', options = options)
def blind_scraping(url) :
    url = url
    driver.get(url)
    time.sleep(0.5)
    titles = driver.find_element_by_xpath('//*[@id="wrap"]/section/div/div[1]/div[1]/h2').text
    main = driver.find_element_by_xpath('//*[@id="contentArea"]').text
    comment = driver.find_element_by_xpath('//*[@id="wrap"]/section/div/div[1]/div[4]').text
    title_list.append(titles)
    main_list.append(main)
    comment_list.append(comment)
    df = pd.DataFrame({'title' : title_list,'main' : main_list,'comment' : comment_list})
    print("정리 끝")
    return df
```

```python
for i in range(0,len(url_df)+1) :
    url = url_df['url'][i]
    please1 = blind_scraping(url)
```

```python
please1['url'] = url_df['url']
please1 = please1[['title','url','main','comment']]
```
