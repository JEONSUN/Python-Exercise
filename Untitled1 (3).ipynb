```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,100)
x
```

```python
y = x ** 2
y
```

```python
plt.style.use('seaborn-whitegrid')
```

```python
plt.plot(x,y)
plt.title('Title')
plt.show()
```

```python
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))

plt.show()
```

```python
fig, ax = plt.subplots(2)

ax[0].plot(x,np.sin(x), color = 'g')
ax[1].plot(x, np.cos(x), color = 'r')
plt.show()
```
