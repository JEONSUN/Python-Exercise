```python
root = "D:/Portfolio/Python Exercise/titanic"
train = pd.read_csv('./train.csv')
test = pd.read_csv('./test.csv')
gender_submission = pd.read_csv('./gender_submission.csv')
```

```python
train = train.drop(['PassengerId', 'Name', 'Ticket','Cabin'],axis=1)
test = test.drop(['Name', 'Ticket','Cabin'],axis=1)
```

```python
sex = {"male":1, "female":0}
train['Sex'] = train['Sex'].map(sex)
test['Sex'] = test['Sex'].map(sex)
embarked = {"S":2, "Q":1, "C":0}
train['Embarked'] = train['Embarked'].map(embarked)
test['Embarked'] = test['Embarked'].map(embarked)
```

```python
grid = sns.FacetGrid(train, row='Pclass', col='Sex', height=2.2, aspect=1.6)
grid.map(plt.hist, 'Age', alpha=0.5, bins=20)
grid.add_legend()
```

```python
train.loc[(train.Age.isnull())&(train.Pclass==1),'Age']=38
train.loc[(train.Age.isnull())&(train.Pclass==2),'Age']=28
train.loc[(train.Age.isnull())&(train.Pclass==3),'Age']=20
train.loc[(train.Embarked.isnull()),'Embarked']=2
train.Age.isnull().any()
train.Embarked.isnull().any()
test.loc[(test.Age.isnull())&(test.Pclass==1),'Age']=38
test.loc[(test.Age.isnull())&(test.Pclass==2),'Age']=28
test.loc[(test.Age.isnull())&(test.Pclass==3),'Age']=20
test.loc[(test.Embarked.isnull()),'Embarked']=2
test.loc[(test.Fare.isnull()),'Fare']=2
test.Age.isnull().any()
test.Embarked.isnull().any()
test.Fare.isnull().any()
```

```python
train.info()
test.info()
```

```python
train["Age"].fillna(train.groupby("Pclass")["Age"].transform("median"), inplace=True)
test["Age"].fillna(test.groupby("Pclass")["Age"].transform("median"), inplace=True)

scaler = StandardScaler()
array = train['Age'].values.reshape(-1,1)
scaler.fit(array)
train['AgeScale'] = pd.Series(scaler.transform(array).reshape(-1))
array = test['Fare'].values.reshape(-1,1)
scaler.fit(array)
test['AgeScale'] = pd.Series(scaler.transform(array).reshape(-1))
```

```python
train["Fare"].fillna(train.groupby("Pclass")["Fare"].transform("median"), inplace=True)
test["Fare"].fillna(test.groupby("Pclass")["Fare"].transform("median"), inplace=True)

scaler = StandardScaler()
array = train['Fare'].values.reshape(-1,1)
scaler.fit(array)
train['FareScale'] = pd.Series(scaler.transform(array).reshape(-1))
array = test['Fare'].values.reshape(-1,1)
scaler.fit(array)
test['FareScale'] = pd.Series(scaler.transform(array).reshape(-1))
```

```python
train = train.drop(['Fare'], axis=1)
test = test.drop(['Fare'], axis=1)
train = train.drop(['Age'], axis=1)
test = test.drop(['Age'], axis=1)
```

```python
train['Family']=1
train['Family']=train['Parch']+train['SibSp']
train = train.drop(['SibSp','Parch'],axis=1)
test['Family']=1
test['Family']=test['Parch']+test['SibSp']
test = test.drop(['SibSp','Parch'],axis=1)
```

```python
train['Alone'] = 0
test['Alone'] = 0
```

```python
for idx, i in enumerate(train['Family']) :
    if i == 1 :
        train.loc[idx,'Alone'] = 1
    else :
        train.loc[idx,'Alone'] = 0
```

```python
for idx, i in enumerate(test['Family']) :
    if i == 1 :
        test.loc[idx,'Alone'] = 1
    else :
        test.loc[idx,'Alone'] = 0
```

```python
# test_data를 복사 (원본 유지용)
test1 = test.copy()
# test data를 기존의 학습 데이터 형식과 맞춤
test1 = test1.drop(["PassengerId"], axis = 1)
```

```python
dataset = train.values
X = dataset[:,1:8].astype(float)
Y = dataset[:,0].astype(float)
trainx, testx, trainy, testy = train_test_split(X,Y,test_size=0.2,random_state=42)

model = Sequential()
model.add(Dense(891, input_dim=7, activation='relu'))
model.add(Dense(891, activation='relu'))
model.add(Dense(1))
```

```python
model.compile(loss='mean_squared_error', optimizer='adam', metrics = ['accuracy'])
model.fit(trainx, trainy, epochs=130, batch_size=10)
expect1 = model.predict(test1).flatten
```

```python
model1 = Sequential()
model1.add(Dense(891, input_dim=7, activation='relu'))
model1.add(Dense(891, activation='relu'))
model1.add(Dense(1, activation = 'sigmoid'))
```

```python
model1.compile(loss='mean_squared_error', optimizer='adam', metrics = ['accuracy'])
model1.fit(trainx, trainy, epochs=130, batch_size=10)
expect2 = model1.predict(test1).flatten
```
